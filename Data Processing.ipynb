```python
mapped = np.apply_along_axis(lambda row: list(row[0]), 1, data)
```

```python
def generate_variables(row):
    n = len(row)
    map_letters = dict({'A' : 0, 'T' : 1, 'C' : 2, 'G' : 3})
    variables = np.zeros([4*n])
    for i in range(n):
        variables[4*i + map_letters[row[i]]] = 1
    return variables
```

```python
variables = np.apply_along_axis(generate_variables, 1, mapped)
```
